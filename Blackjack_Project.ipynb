### Rules:
`1. ` The game will have two players: the Dealer and the Player. The game will start off with a deck of 52 cards. The 52 cards will consist of 4 different suits: Clubs, Diamonds, Hearts and Spades. For each suit, there will be cards numbered 1 through 13. <br>

**Note: No wildcards will be used in the program**

`2. ` When the game begins, the dealer will shuffle the deck of cards, making them randomized. After the dealer shuffles, it will deal the player 2 cards and will deal itself 2 cards from. The Player should be able to see both of their own cards, but should only be able to see one of the Dealer's cards.
 
`3. ` The objective of the game is for the Player to count their cards after they're dealt. If they're not satisfied with the number, they have the ability to 'Hit'. A hit allows the dealer to deal the Player one additional card. The Player can hit as many times as they'd like as long as they don't 'Bust'. A bust is when the Player is dealt cards that total more than 21.

`4. ` If the dealer deals the Player cards equal to 21 on the **first** deal, the Player wins. This is referred to as Blackjack. Blackjack is **NOT** the same as getting cards that equal up to 21 after the first deal. Blackjack can only be attained on the first deal.

`5. ` The Player will never see the Dealer's hand until the Player chooses to 'stand'. A Stand is when the player tells the dealer to not deal it anymore cards. Once the player chooses to Stand, the Player and the Dealer will compare their hands. Whoever has the higher number wins. Keep in mind that the Dealer can also bust.

```python
from random import shuffle
from IPython.display import clear_output

class Deck:
    def __init__(self):
        self.suits = ['Clubs', 'Diamonds', 'Hearts', 'Spades']
        self.numbers = list(range(1, 14))
        self.deck = self.get_deck()
        self.shuffle_deck()

    def get_deck(self):
        return [{'suit': suit, 'number': number} for suit in self.suits for number in self.numbers]

    def shuffle_deck(self):
        shuffle(self.deck)

class Player:
    def __init__(self):
        self.hand = []
        self.tally = 0

    def get_score(self):
        return sum(card['number'] for card in self.hand)

    def display_bust_message(self):
        print("Bust! You've exceeded 21.")

    def display_hand(self):
        print("Your current hand:")
        for card in self.hand:
            print(f"{card['number']} of {card['suit']}")

    def display_blackjack_message(self):
        print("Blackjack! You've got 21.")

    def display_win_message(self):
        print("You win!")

class Human(Player):
    def __init__(self):
        super().__init__()

    def display_human(self):
        print("Human Player")

    def give_hit(self, deck):
        self.hand.append(deck.pop())


class Dealer(Player):
    def __init__(self):
        super().__init__()

    def deal_hand(self, deck):
        self.hand = [deck.pop(), deck.pop()]

    def give_hit(self, deck):
        self.hand.append(deck.pop())

    def tally_and_show_hand(self,deck):
        while self.get_score() < 16:
            self.give_hit(deck)
        print("Dealer's hand:")
        for card in self.hand:
            print(f"{card['number']} of {card['suit']}")

    def display_dealer(self):
        print("Dealer Player")

class Game:
    def __init__(self):
        self.deck = Deck()
        self.dealer = Dealer()
        self.human = Human()
        self.players = [self.dealer, self.human]

    def display_win_message(self):
        print("You win!")

    def display_push_message(self):
        print("It's a push! The game is tied.")

    def display_loss_message(self):
        print("You lose!")

    def main(self):
        # Ask the player how many decks they want to use
        while True:
            try:
                num_decks = int(input("Enter the number of decks to use: "))
                if num_decks > 0:
                    break  # Break the loop if a valid integer is provided
                else:
                    print("Please enter a positive number of decks.")
            except ValueError:
                print("Invalid input. Please enter a valid number.")

        print(f"Using {num_decks} decks.")

        # Create and shuffle the deck
        self.deck = Deck()
        self.deck.deck *= num_decks
        self.deck.shuffle_deck()

        # Game loop
        while True:
            # Display hands
            self.human.display_hand()
            self.dealer.display_dealer()

            # Check for Blackjack
            if self.human.get_score() == 21:
                self.human.display_blackjack_message()
                break

            # Ask the player if they want a hit
            hit_choice = input("Do you want a hit? (yes/no): ").lower()
            if hit_choice == 'yes':
                self.human.give_hit(self.deck.deck)

                # Check for bust
                if self.human.get_score() > 21:
                    self.human.display_hand()
                    self.human.display_bust_message()
                    break
            else:
                # Dealer's turn
                self.dealer.tally_and_show_hand(self.deck.deck)


                # Compare scores and display results
                if self.human.get_score() > self.dealer.get_score():
                    self.display_win_message()
                elif self.human.get_score() == self.dealer.get_score():
                    self.display_push_message()
                else:
                    self.display_loss_message()

                break

                
def main():
        game_instance = Game()
        game_instance.main()
```
